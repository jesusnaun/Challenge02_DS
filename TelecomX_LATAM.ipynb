#ðŸ”§ TransformaciÃ³n

```python
customer = pd.json_normalize(clientes["customer"])
customer
```

```python
phone = pd.json_normalize(clientes["phone"])
phone
```

```python
internet = pd.json_normalize(clientes["internet"])
internet
```

```python
account = pd.json_normalize(clientes["account"])
account
```

```python
columnas_a_eliminar = ["customer","phone","internet","account"]
clientes.drop(columns=columnas_a_eliminar, inplace=True)
clientes
```

```python
datos_normalizados = pd.concat([clientes, customer, phone, internet, account], axis=1)
```

```python
datos_normalizados.head()
```

```python
datos_normalizados.dtypes
```

```python
datos_normalizados.info()
```

```python
#columnas=list(datos_normalizados[["TechSupport","StreamingTV"]])
columnas=list(datos_normalizados.columns)
for col in columnas:
  print(col)
  print(datos_normalizados[col].unique())
```

```python
unique_values = datos_normalizados["customerID"].unique
unique_values
```

```python
datos_normalizados["customerID"] = datos_normalizados["customerID"].astype("string")
```

```python
datos_normalizados.info()
```

```python
datos_normalizados["Churn"] = datos_normalizados["Churn"].replace({"Yes": 1, "No": 0, "": 0})
datos_normalizados
```

```python
#datos_normalizados['Churn'] = pd.to_numeric(datos_normalizados['Churn'])

datos_normalizados['Churn'] = datos_normalizados['Churn'].astype("int8")
```

```python
# Agrupar por tenure y Churn
df_grouped = datos_normalizados.groupby(['tenure', 'Churn']).size().unstack(fill_value=0)

# Calcular el % de churn por tenure
df_grouped['churn_rate'] = (df_grouped.get('Yes', 0) / df_grouped.sum(axis=1)) * 100

# Mostrar el resultado
print(df_grouped[['churn_rate']])
```

```python
fig = px.line(df_grouped.reset_index(),
              x='tenure',
              y='churn_rate',
              markers=True,
              title='Tasa de Churn (%) por Tenure')
fig.show()
```
